Now we want to create databunch.

```
data = (ImageList.from_folder('/root/data/mammography_full_dataset/dataset')
        .split_by_rand_pct()
        .label_from_folder()
        .transform(tfms=get_transforms(), size=224)
        .databunch()
        .normalize(imagenet_stats))
```

```
data.save(root_dir + 'models/mamography_models/databunch.pkl')
```

```
data = load_data(root_dir + 'models/mamography_models', 'databunch.pkl', bs=bs)
```

```
data.classes, len(data.train_ds), len(data.valid_ds)
```

###Training learner

Now we create learner

```
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

Find and plot lr

```
learn.lr_find()
```

```
learn.recorder.plot()
```

Train top levels

```
learn.fit_one_cycle(5, 3e-3)
learn.save(root_dir + 'models/mamography_models/stage-1-1')
```

```
learn.load(root_dir + 'models/mamography_models/stage-1-1')
```

Unfreeze the entire thing.

```
learn.unfreeze()
```

Now we train the entire thing.

```
learn.fit_one_cycle(2, max_lr=slice(3e-6,3e-4))
learn.save(root_dir + 'models/mamography_models/stage-2')
```

```
learn.fit_one_cycle(2, max_lr=slice(3e-6,3e-4))
learn.save(root_dir + 'models/mamography_models/final_model')
```
